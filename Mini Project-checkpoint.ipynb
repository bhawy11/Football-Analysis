### Add two new columns

```python
epl_data['MinsPerMatch']=(epl_data['Mins']/epl_data['Matches']).astype(int)
epl_data['GoalsPerMatch']=(epl_data['Goals']/epl_data['Matches']).astype(float)
epl_data.head()
                           
```

```python
Total_Goals=epl_data['Goals'].sum()
print(Total_Goals)
```

```python
Total_PenaltyGoals=epl_data['Penalty_Goals'].sum()
print(Total_PenaltyGoals)
```

```python
Total_Penalty_Received=epl_data['Penalty_Attempted'].sum()
print(Total_Penalty_Received)
```

## Pie chart for penalties missed vs scored


```python
plt.figure(figsize = (13,6))
penalty_not_scored = epl_data['Penalty_Attempted'].sum() - Total_PenaltyGoals
data = [penalty_not_scored, Total_PenaltyGoals]
labels = ['Penalties Misses','Penalties Scored']
color_palette=sns.color_palette("Paired")
plt.pie(data,labels=labels,colors=color_palette,autopct='%.0f%%')
plt.show()
```

## Total Assist from each club

```python
Assists_by_club=pd.DataFrame(epl_data.groupby('Club', as_index=False)['Assists'].sum())
sns.set_theme(style='whitegrid',color_codes=True)
ax=sns.barplot(x='Club',y='Assists', data=Assists_by_club.sort_values(by='Assists'),palette='tab20')
ax.set_xlabel("Club",fontsize=30)
ax.set_ylabel("Assists",fontsize=20)
plt.xticks(rotation=75)
plt.rcParams['figure.figsize']=(20,8)
plt.title('Plot of Club vs Total Assists',fontsize=20)
```

```python
top_10_assists=epl_data[['Name','Club','Assists','Matches']].nlargest(n=10,columns='Assists')
top_10_assists
```

```python
Goals_by_clubs = pd.DataFrame(epl_data.groupby('Club', as_index = False)['Goals'].sum())
sns.set_theme(style ="whitegrid", color_codes = True)
ax = sns.barplot(x = 'Club', y = 'Goals', data = Goals_by_clubs.sort_values(by ="Goals"), palette = 'rocket')
ax.set_xlabel("Club", fontsize = 30)
ax.set_ylabel("Goals", fontsize = 20)
plt.xticks(rotation =75)
plt.rcParams["figure.figsize"] = (20, 8)
plt.title('Plot of Club vs Total Goals', fontsize = 20)
```

```python
top_10_goals = epl_data[['Name', 'Club', 'Goals', 'Matches']].nlargest(n = 10, columns = 'Goals')
top_10_goals
```

```python
plt.figure(figsize = (14, 7))
assists = epl_data['Assists'].sum()
data = [Total_Goals - assists, assists]
labels = ['Goals without assists', 'Goals with assists']
color = sns.color_palette('Set2')
plt.pie(data, labels = labels, colors = color, autopct ='%.0f%%')
plt.title('Percentage of Goals with Assists', fontsize = 20)
plt.show()
```

```python
epl_yellow = epl_data.sort_values(by = 'Yellow_Cards', ascending = False)[:10]
plt.figure(figsize = (20, 6))
plt.title("Players with the most yellow cards")
c = sns.barplot(x = epl_yellow['Name'], y = epl_yellow['Yellow_Cards'], label = 'Players', color ='yellow')
plt.ylabel('Number of Yellow Cards')
c.set_xticklabels(c.get_xticklabels(), rotation = 45)
plt.show()
```

```python
epl_yellow = epl_data.sort_values(by = 'Red_Cards', ascending = False)[:10]
plt.figure(figsize = (20, 6))
plt.title("Players with the most Red cards")
c = sns.barplot(x = epl_yellow['Name'], y = epl_yellow['Red_Cards'], label = 'Players', color ='Red')
plt.ylabel('Number of Red Cards')
c.set_xticklabels(c.get_xticklabels(), rotation = 45)
plt.show()
```
